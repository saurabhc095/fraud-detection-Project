```python
## Step 1: Load Libraries and Dataset

# Import essential libraries for data handling and visualization
import pandas as pd                  # For data manipulation
import matplotlib.pyplot as plt      # For basic plotting
import seaborn as sns                # For advanced statistical plots
import plotly.express as px          # For interactive plots
from collections import Counter      # For counting elements
```

```python
# Load the dataset from the specified file path
file_path = r"C:\Users\saura\OneDrive\Desktop\DATA S\New project\dataset\creditcard.csv"
df = pd.read_csv(file_path)         # Read CSV file into DataFrame
df.head()                           # Display first 5 rows of the dataset
```

```python
## Step 2: Basic Dataset Info
print(df.shape)     # Print number of rows and columns in the dataset
```

```python
print(df.info())    # Print column data types and non-null counts
```

```python
df.describe()       # Summary statistics of numeric columns
```

```python
## Step 3: Class Distribution
# Count the number of occurrences in the 'Class' column
class_counts = df['Class'].value_counts()

# Calculate the percentage distribution of classes
percent = df['Class'].value_counts(normalize=True) * 100
```

```python
# Plot class distribution using seaborn
sns.countplot(data=df, x='Class', palette='Set2')
plt.title("Class Distribution")
plt.show()
```

```python
# Print the percentage distribution
print("Class Distribution (%):")
print(percent)
```

```python
## Step 4: Check for Duplicate Rows

duplicate_rows = df.duplicated().sum()  # Count number of duplicate rows
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
## Step 5: Correlation Analysis

# Calculate correlation of each feature with 'Class' (absolute values)
correlation = df.corr()['Class'].abs().sort_values(ascending=False)

# Print top 10 features most correlated with the target
print("Top features correlated with Class:")
print(correlation.head(10))
```

```python
#Step 6: Feature Engineering

# Create 'Hour' feature from 'Time' column (convert seconds to hour of day)
df['Hour'] = df['Time'].apply(lambda x: int(x // 3600) % 24)
```

```python
# Create 'Log_Amount' by applying log transformation on 'Amount'
import numpy as np
df['Log_Amount'] = np.log1p(df['Amount'])  # log1p handles zero safely
```

```python
# Step 7: Save Cleaned Data for Week 2

# Drop any duplicate rows before saving
df = df.drop_duplicates()
```

```python
# Save the cleaned data to a new CSV file for use in Week 2
df.to_csv("processed_data.csv", index=False)
```
